```python
import numpy as np
from scipy. integrate import odeint
import matplotlib.pyplot as plt
import math

t0 = 0
x0 = 24
N = 3030

a1 = 0.288
a2 = 0.000018

t = np.arange( t0, 20, 0.1)

def syst(dx, t):
    x = dx
    return (a1 +x*a2)*(N-x)


y = odeint(syst, x0, t)

fig1 = plt.figure(facecolor='white')
plt.plot(t, y, linewidth=2, label="")
plt.xlabel("t")
plt.ylabel("")
plt.grid(True)
plt.legend()
plt.show()
fig1.savefig('03.png', dpi = 600)

a1 = 0.000018
a2 = 0.377

t = np.arange( t0, 0.15, 0.001)

y = odeint(syst, x0, t)
dy = (a1 +y*a2)*(N-y)

fig2 = plt.figure(facecolor='white')
plt.plot(t, y, linewidth=2, label="")
plt.plot(t, dy, linewidth=2, label="")
plt.xlabel("t")
plt.ylabel("")
plt.grid(True)
plt.legend()
plt.show()
fig2.savefig('04.png', dpi = 600)

def a1(t): 
    a1 = 0.1*t
    return a1

def a2(t): 
    a2 = 0.4*math.cos(t)
    return a2

t = np.arange( t0, 0.02, 0.001)

def syst2(dx, t):
    x = dx
    return (a1(t) +x*a2(t))*(N-x)

y = odeint(syst2, x0, t)

fig3 = plt.figure(facecolor='white')
plt.plot(t, y, linewidth=2, label="")
plt.xlabel("t")
plt.ylabel("")
plt.grid(True)
plt.legend()
plt.show()
fig3.savefig('05.png', dpi = 600)
```
